# Logistic regression for a 2D classification problem

Imports...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Packages for logistic regression: scikit-learn or statsmodels
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
```

## Load and visualize data

```python
data = pd.read_csv("2d-classification-data.csv")
data.head(10)
```

```python
data.plot.scatter(x="z1", y="z2", c="y");
```

## Define and fit the model

```python
# Select model input variables
# TODO
X = data[['z1', 'z2']]
y = data['y']
```

```python
# Define a model and fit it
# Note: sklearn.linear_model.LogisticRegression adds an intercept automatically.
# TODO
log_reg = LogisticRegression()
log_reg.fit(X, y)
```

```python
log_reg.intercept_ # ß_0
```

```python
log_reg.coef_ # ß_1 & ß_2
```

## Use model for Prediction

```python
# Prediction of training samples
# TODO
predicted_classes = log_reg.predict(X)
print("Predicted classes: ", predicted_classes)
predicted_probabilities = log_reg.predict_proba(X)
print("Predicted probabilities: ", predicted_probabilities)
```

```python
# Parameters
# TODO
print("Beta zero: ", log_reg.intercept_)
print("Beta one and beta two: ", log_reg.coef_)
```

## Visualization

```python
# Predictions with decision boundary
# TODO

# Define the range of z1 and z2 based on the data
x1_min, x1_max = data['x1'].min() - 1, data['x1'].max() + 1
x2_min, x2_max = data['x2'].min() - 1, data['x2'].max() + 1

# Create a mesh grid of points
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max, 100), np.linspace(x2_min, x2_max, 100))

# Calculate quadratic terms for each point in the grid
z1 = xx1 ** 2
z2 = xx2 ** 2

# Flatten the grid arrays for prediction
grid_points = np.c_[z1.ravel(), z2.ravel()]
```

```python
# Decision boundary vs. ground truth
# TODO
# Predict the probabilities for each point in the grid
probs = log_reg.predict_proba(grid_points)[:, 1]  # Probability of class 1
probs = probs.reshape(xx1.shape)
```

```python
# Probabilities
# TODO (optional)
# Plot decision boundary
plt.figure(figsize=(10, 6))

# Plot the contour where the probability of class 1 is 0.5
contour = plt.contourf(xx1, xx2, probs, levels=[0, 0.5, 1], alpha=0.3, colors=['blue', 'red'])
plt.colorbar(contour)

# Plot the original data points
plt.scatter(data['x1'], data['x2'], c=data['y'], cmap='bwr', edgecolor='k', s=50)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Decision Boundary of Logistic Regression Model')
plt.show()
```

```python
#d)
# Define the range for x1 and x2 based on the data
x1_min, x1_max = data['x1'].min() - 1, data['x1'].max() + 1
x2_min, x2_max = data['x2'].min() - 1, data['x2'].max() + 1

# Create a mesh grid
xx1, xx2 = np.meshgrid(np.linspace(x1_min, x1_max, 100), np.linspace(x2_min, x2_max, 100))

# Compute the quadratic terms for the mesh grid
z1 = xx1 ** 2
z2 = xx2 ** 2
grid_points = np.c_[z1.ravel(), z2.ravel()]

# Predict probabilities using the logistic regression model
probs = log_reg.predict_proba(grid_points)[:, 1]
probs = probs.reshape(xx1.shape)

# Plot the decision boundaries
plt.figure(figsize=(10, 6))

# Plot the logistic regression model's decision boundary
contour = plt.contourf(xx1, xx2, probs, levels=[0, 0.5, 1], alpha=0.3, colors=['blue', 'red'])
plt.colorbar(contour)

# Plot the actual circular decision boundary
theta = np.linspace(0, 2 * np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
plt.plot(circle_x, circle_y, 'k--', label='Actual Decision Boundary (radius=1)')

# Plot the original data points
plt.scatter(data['x1'], data['x2'], c=data['y'], cmap='bwr', edgecolor='k', s=50)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Comparison of Logistic Regression Model vs Actual Decision Boundary')
plt.legend()
plt.show()
```
