## Imports and data loading

````python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
````

````python
data = pd.read_csv("gauss-markov.csv")
data.head(10)
````

## a)

We start by using the simple linear regression model
$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3
$$

Using [``sm.OLS``](https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.OLS.html), compute optimal values for the parameters.

> Note: You may want to use ``sm.add_constant`` to add values for the intercept.



````python
data.filter(["x1", "x2", "x3"])
````

````python
# Prepare input data
X = data[["x1", "x2", "x3"]]
y = data["y"]
````

````python
# Fit a linear model with statsmodels
X = sm.add_constant(X) # first we have to add a constant term to the independent values
model = sm.OLS(y, X) 
results = model.fit()
````

````python
# Show the results using the summary() function
print(results.summary())
````

Use the model to predict the $y$-values

````python
predicted_values = results.predict(X) 
````

````python
predicted_values
````

````python
# Visualization of the predicted variables vs. the true variables
# NOTHING TO DO here
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
for ax, variable_name in zip(axs, ["x1", "x2", "x3"]):
    ax.scatter(data[variable_name], data["y"], label="Ground truth")
    ax.scatter(data[variable_name], predicted_values, label="Model prediction")
    ax.legend()
    ax.set_xlabel(variable_name)
````

## b)

Compute the residuals $e = \hat{y} - y$ of the resulting model.

````python
residuals = results.resid
````

````python
residuals
````

Plot the residuals over the input variables $x_1$ and $x_2$. What do you observe?

````python
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(data["x1"], residuals)
plt.xlabel("x1")
plt.ylabel("residuals")
plt.title("Residuals vs x1")

plt.subplot(122)
plt.scatter(data["x2"], residuals)
plt.xlabel("x2")
plt.ylabel("residuals")
plt.title("Residuals vs x2")

plt.tight_layout()
plt.plot()
````

- We notice that the residuals over x1 are have a V-shape or curved pattern, indicating **non-linearity**. This indicates a non-linear term should be added to the model.
- The residuals over x2 are randomly scattered with no pattern. It suggests that the linear model assumptions are likely met.

Using a White test ([`statsmodels.stats.diagnostic.het_white`](https://www.statsmodels.org/dev/generated/statsmodels.stats.diagnostic.het_white.html)), show that the linear model does not satisfy homoscedasticity.

````python
from statsmodels.stats.diagnostic import het_white

white_test = het_white(residuals, X)

print("White's test for heteroscedasticity")
print(f"LM statistic {white_test[0]:.4f}")
print(f"LM p-value {white_test[1]:.4f}")
print(f"F statistic {white_test[2]:.4f}")
print(f"F p-value {white_test[3]:.4f}")
````

- We assume that we have a significance level of $\alpha=0.01$
- the $p$-values of the White test is 0.
- Since $p$<$\alpha$, we reject the Null Hypothesis.
- This suggests the presence of **heteroscedasticity**.
- If the $p$-values would be greater than our significane level, we fail to reject the Null Hypothesis which suggests **homoscedasticity**.

## c)

Consider the alternative model
$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \beta_3 X_3 + \beta_4 X_1^2
$$

Compute the optimal parameter values. You should observe that the $R^2$ value improves drastically over the previous model.

````python
# Prepare input data
X = X[["x1", "x2", "x3"]]
X["x1_squared"] = X["x1"] ** 2
y = data["y"]
X = sm.add_constant(X) # add a constant term to the independent values
# Fit a linear model
model_new = sm.OLS(y, X) 
results_new = model_new.fit()
````

````python
print(results_new.summary())
````

````python
print(f"R-squared (original model): {results.rsquared:.4f}")
print(f"R-squared (new model with x1^2): {results_new.rsquared:.4f}")
````

Although this model gives a very good fit of the data, there is another problem: Multicolliniearity.
Use the Variance inflation factor ([`statsmodels.stats.outliers_influence.variance_inflation_factor`](https://www.statsmodels.org/dev/generated/statsmodels.stats.outliers_influence.variance_inflation_factor.html)) to check whether the variables are dependent.

````python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [vif(X.values, i) for i in range(X.shape[1])]

print("Variance influence factors:")
print(vif_data)
````

- We observe very high VIF values for $x_1$ and $x_1$ squared.
- Interpreting the results:
    VIF values greater than 1 indicate some correlation between variables.
    VIF values between 5 and 10 suggest high correlation that may be problematic.
    VIF values above 10 indicate severe multicollinearity.
