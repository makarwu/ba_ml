b) Plot a histogram of the attribute "age". Which is the most frequent age?

```python
plt.hist(df_new['age'])
print("most frequent age", df_new['age'].value_counts())
# 39 is the age with the highest frequency
```

- c) Plot the average number of "kids" against "age" and interpret the resulting graph.
Compute the correlation between "kids" and "age" to check your interpretation.

```python
plt.scatter(df_new['age'], df_new['kids'])
print("correlation between kids and age is:", df_new['kids'].corr(df_new['age']))
```

- d) Plot "log of hourly wage (lnwg)" against "age".

```python
plt.plot(df_new['age'], df_new['lnwg'])
```

- e) Plot the mean of "log of hourly wage (lnwg)" against "age".

```python
plt.scatter(df_new['age'], df_new['lnwg'])
print(f"the correlation between lnwg and age is: {df_new['lnwg'].corr(df_new['age'])}")
# There is almost no correlation between those two attributes
```
